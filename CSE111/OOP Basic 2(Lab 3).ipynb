```python
#TASK 5
class Shape():
    def __init__(self,shape,x,y):
        self.x = x
        self.y = y
        self.shape = shape
        self.areaa = 0
    def area(self):
        if self.shape == "Triangle" or self.shape == "Rhombus":
            self.area = (self.x * self.y)/2
        elif self.shape == "Square" or self.shape == "Rectangle":
            self.area = self.x * self.y
        else:
            self.area = "Shape unknown"
        print("Area:", self.area)

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#TASK 6
class Calculator():
    def __init__(self):
        print("Calculator is ready!")
    def calculate(self,x,y,sign):
        self.x = x
        self.y = y
        self.sign = sign
        if self.sign == "+":
            self.val = self.x + self.y
        elif self.sign == "-":
            self.val = self.x - self.y
        elif self.sign == "*":
            self.val = self.x * self.y
        else:
            self.val = self.x / self.y
        return self.val
    def showCalculation(self):
        print(self.x, self.sign, self.y) 

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#TASK 7
class Student():
    def __init__(self,name,ID,department,cg):
        self.name = name
        self.id = ID
        self.department = department
        self.cg = cg
    def calculate_CGPA(self):
        temp = 0
        for i in self.cg:
            temp += i * 3
        self.CGPA = temp/(len(self.cg)*3)
    def print_details(self):
        print(f"Name: {self.name}, ID: {self.id}\nDepartment: {self.department}")
        print("CGPA:", self.CGPA)
        if self.CGPA > 3.8:
            print("Your academic standing is 'Highest Distinction'")
        elif self.CGPA > 3.65:
            print("Your academic standing is 'High Distinction'")
        elif self.CGPA > 3.5:
            print("Your academic standing is 'Distinction'")
        elif self.CGPA > 2:
            print("Your academic standing is 'Satisfactory'")
        else:
            print("Sorry, you cannot graduate")

s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()

```

```python
#TASK 8
class Shinobi:
    def __init__(self,name,rank):
        self.name = name
        self.rank = rank
        self.salary = 0
        self.mission = 0
    def calSalary(self,mission):
        self.mission = mission
        if self.rank == "Genin":
            self.salary = self.mission * 50
        elif self.rank == "Chunin":
            self.salary = self.mission * 100
        else:
            self.salary = self.mission * 500
    def printInfo(self):
        print(f"Name: {self.name}\nRank: {self.rank}\nNumber of mission: {self.mission}\nSalary: {self.salary}")
    def changeRank(self,rank):
        self.rank = rank

naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()

```

```python
#TASK 10
class UberEats:
    def __init__(self,name,phone,address):
        self.name = name
        self.phone = phone
        self.address = address
        print(f"{self.name}, welcome to UberEats!")
    def add_items(self,itm1,itm2,p1,p2):
        self.order = {}
        self.order[itm1] = p1
        self.order[itm2] = p2
        self.price = p1 + p2
    def print_order_detail(self):
        temp = "User details: Name: " + self.name + ", Phone: " + self.phone + ", Address: " + self.address + "\nOrders: " + str(self.order) + "\nTotal Paid Amount: " + str(self.price)
        return temp

order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```

```python
#TASK 11
class Spotify:
    def __init__(self,songlist):
        print("Welcome to Spotify!")
        self.songlist = songlist
    def playing_number(self,num):
        self.num = num
        if  self.num > len(self.songlist):
            temp1 = str(self.num) + " number song not found. Your playlist has " + str(len(self.songlist)) + " songs only"
            return temp1
        else:
            temp2 = "##########################" + "\nPlaying " + str(self.num) + " number song for you\nSong name: " + str(self.songlist[self.num-1])
            return temp2
    def add_to_playlist(self,song):
        self.song = song
        self.songlist.append(self.song)

user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
