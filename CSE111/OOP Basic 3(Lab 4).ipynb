```python
#TASK 7
class Match:
  def __init__(self,match):
    list1 = match.split("-")
    self.batTeam = list1[0]
    self.bowlTeam = list1[1]
    self.over = 0
    self.run = 1
    self.wicket = 0
    print("5..4..3..2..1.. Play !!!")
  def add_run(self,run):    
    self.run += run
  def add_over(self,over):
    if over >= 5:
      print("Warning! Cannot add 5 over/s (5 over match)")
    else:
      self.over += over
  def add_wicket(self,wicket):
    self.wicket += wicket
  def print_scoreboard(self):
    temp = "Batting Team: " + self.batTeam + "\nBowling Team: " + self.bowlTeam + "\nTotal Runs: " + str(self.run) + " Wickets: " + str(self.wicket) + " Overs: " + str(self.over)
    return temp
match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
#TASK 14
class StudentDatabase:
  def __init__(self,name,id):
    self.name = name
    self.id = id
    self.grades = {}
  def calculateGPA(self,list1,semester):
    addedgrade = 0
    list2 = []
    for i in list1:
      temp = i.split(" ")
      addedgrade += float(temp[1])
      list2.append(temp[0])
    grade = addedgrade/len(list1)
    tuple1 = tuple(list2)
    mydic = {}
    mydic[tuple1] = grade
    self.grades[semester] = mydic
  def printDetails(self):
    print(f"Name: {self.name}\nID: {self.id}")
    for k1 in self.grades.keys():
      print(f"Courses taken in {k1}")
      for v1 in self.grades.values():
        for k2 in v1.keys():
          for i in k2:
            print(i[:-1])
        for v2 in v1.values():
          print("GPA:",round(v2,2))

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
