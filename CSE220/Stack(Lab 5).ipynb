```python
class ArrayStack:
  def __init__(self,string=""):
    self.array = []
    for i in string:
      self.array.append(i)
  def push(self,p):
    self.array.append(p)
  def pop(self):
    if len(self.array) == 0:
      return None
    lastElement = self.array[-1]
    del self.array[-1]
    return lastElement
  def peak(self):
    if len(self.array) == 0:
      return None
    return (self.array[-1])
```

```python
class Node:
  def __init__(self,elem,next):
    self.element = elem
    self.next = next

class LinkedListStack:
  def __init__(self):
    self.head = None
  def push(self,p):
    if self.head == None:
      self.head = Node(p,None)
    else:
      n = Node(p,self.head)
      self.head = n
  def pop(self):
    if self.head == None:
      return None
    else:
      poppedElement = self.head.element
      poppedNode = self.head
      self.head = self.head.next
      poppedNode.element = 0
      poppedNode.next = 0
      return poppedElement
  def peak(self):
    if self.head == None:
      return None
    else:
      return self.head.element
```

```python
#Task 1
stringval = input()
a = ArrayStack()
flag = True
tempList = []
prefixCount = 0
index = 0
temp = ""
for i in stringval:
  prefixCount += 1
  if i in "[{(":
    a.push(i)
    tempList.append(prefixCount)
  elif i in ")}]":
    lastelement = a.peak()
    if chr(ord(i)-1) == lastelement or chr(ord(i)-2) == lastelement:
      a.pop()
      del tempList[-1]
    elif lastelement == None:
      index = prefixCount
      temp = i
      flag = False
      break
    else:
      flag = False
      if tempList != []:
        index = tempList[-1]
      break
  else:
    pass
if a.peak() != None:
  flag = False
  if index == 0:
    index = tempList[-1]
if flag == True:
  print("This expression is correct.")
else:
  print("This expression is NOT correct.")
  if index != 0:
    if temp == "":
      print(f"Error at character # {index}.'{a.peak()}'- not closed.")
    else:
      print(f"Error at character # {index}.'{temp}'- not opened.")

```
