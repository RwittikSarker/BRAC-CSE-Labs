```python
#Task 1
def maxval(a):
  max = a[0]
  for i in a:
    if i > max:
      max = i
  return max

def minval(a):
  min = a[0]
  for i in a:
    if i < min:
      min = i
  return min

def negtraverse(a):
  for i in a:
    if i < 0:
      return True
  return False

class KeyIndex:
  def __init__(self,a):
    self.marray = a
    self.k = [0]*(maxval(a)+1)
    self.x = 0
    if negtraverse(a) == False:
      for i in a:
        self.k[i] += 1
    else:
      self.x = minval(a) * (-1)
      self.k = self.k + [0]*self.x
      for i in a:
        self.k[i+self.x] += 1
  def search(self,val):
    if (val+self.x) >= len(self.k) or self.k[val+self.x] == 0:
      return False
    else:
      return True
  def sort(self):
    sortedarray = [0]*len(self.marray)
    i = 0
    for j in range(0,len(self.k)):      
      if self.k[j] != 0:
        if self.k[j] == 1:
          sortedarray[i] = j-self.x
          i += 1
        else:
          temp = self.k[j]
          while temp != 0:
            sortedarray[i] = j-self.x
            i += 1
            temp -= 1
    return sortedarray

a = KeyIndex([4,-2,3,-4,7,4])
print(a.search(8))
print(a.sort())
```

```python
#Task 2
strarray = ["ST1E89B8A32","STEIN69BERG","AQZ","TOP","069YEP","GATOUTOFHELL","SAINTSROW4","TATAKAYE","LENOFARRAY1"]
def hashtable(a):
  k = [None]*9
  for i in a:
    hashval = 0
    #consonant = 0
    for j in i:
      if j not in "AEIOU":
        if j in "0123456789":
          hashval += int(j)
        else:
          hashval += 24
          #consonant += 1
    hashval = hashval % 9
    if k[hashval] == None:
      k[hashval] = i
    else:
      for l in range(1,9):
        if k[(hashval+l)%9] == None:
          k[(hashval+l)%9] = i
          break
  return k
print(hashtable(strarray))
```
