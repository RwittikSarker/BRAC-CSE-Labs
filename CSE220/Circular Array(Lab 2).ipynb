```python
class CircularArray:
  def __init__(self, lin, st, sz):
    # Initializing Variables
    self.start = st
    self.size = sz
    self.cir = [0]*len(lin)
    self.lin = lin
    self.end = (st+sz-1)%len(self.cir)

    # if lin = [10, 20, 30, 40, None]
    # then, CircularArray(lin, 2, 4) will generate
    # cir = [40, null, 10, 20, 30]
    
    # To Do. 
    # Hints: set the values for initialized variables
    idx = self.start
    for i in range(0,len(self.lin)):
      self.cir[idx] = self.lin[i]
      idx = (idx+1) % len(self.cir)
  # Print from index 0 to len(cir) - 1
q  def printFullLinear(self): #Easy
    # To Do
    for i in range(0,len(self.cir)):
      if i == len(self.cir)-1:
        print(self.cir[i])
      else:
        print(self.cir[i],end = ", ")
  
  # Print from start index and total size elements
  def printForward(self): #Easy
    # To Do
    idx = self.start
    for i in range(0,self.size):
      if i == self.size-1:
        print(self.cir[idx])
      else:
        print(self.cir[idx],end = ", ")
        idx = (idx+1) % len(self.cir)

  
  def printBackward(self): #Easy
    # To Do
    idx = self.end
    for i in range(0,self.size):
      if i == self.size-1:
        print(self.cir[idx])
      else:
        print(self.cir[idx],end = ", ")
        idx = idx-1
        if idx < 0:
          idx = len(self.cir)-1
  
  # With no null cells
  def linearize(self): #Medium
    # To Do
    for i in range(0,self.start):
      temp = self.cir[0]
      for j in range(0,len(self.cir)-1):
        self.cir[j] = self.cir[j+1]
      self.cir[-1] = temp
    temp_array = [0]*self.size
    for i in range(0,self.size):
      temp_array[i] = self.cir[i]
    self.cir = [0]*self.size
    for i in range(0,self.size):
      self.cir[i] = temp_array[i]  
  
  # Do not change the Start index
  def resizeStartUnchanged(self, newcapacity): #Medium
    # To Do
    temp_array = [None]*newcapacity
    idx1 = self.start
    idx2 = self.start
    for i in range(0,self.size):
      temp_array[idx1] = self.cir[idx2]
      idx1 = (idx1+1)%len(temp_array)
      idx2 = (idx2+1)%len(self.cir)
    self.cir = [None]*newcapacity
    for i in range(0,newcapacity):
      self.cir[i] = temp_array[i]
  
  # This method will check whether the array is palindrome or not
  def palindromeCheck(self): #Hard
    # To Do
    temp_array1 = [0]*self.size
    idx = self.start
    for i in range(0,self.size):
      temp_array1[i] = self.cir[idx]
      idx = (idx+1) % len(self.cir)
    temp_array2 = [0]*self.size
    temp = 0
    for i in range(self.size-1,-1,-1):
      temp_array2[i] = temp_array1[temp]
      temp+=1
    if temp_array1 == temp_array2:
      print("This array is a palindrome")
    else:
      print("This array is NOT a palindrome")


  # This method will sort the values by keeping the start unchanged
  def sort(self):
    # To Do
    for i in range(0,self.size):
      idx = (self.start+1) % len(self.cir)
      for j in range(0,self.size-1):
        if self.cir[idx-1] > self.cir[idx]:
          temp = self.cir[idx-1]
          self.cir[idx-1] = self.cir[idx]
          self.cir[idx] = temp
        idx = (idx+1) % len(self.cir)
    return self.cir
  
  # This method will check the given array across the base array and if they are equivalent interms of values return true, or else return false
  def equivalent(self, cir_arr):
    # To Do
    if self.size != cir_arr.size:
      return False
    idx1 = cir_arr.start
    idx2 = self.start
    for i in range(0,self.size):
      if self.cir[idx2] != cir_arr.cir[idx1]:
        return False
      idx2 = (idx2+1) % len(self.cir)
      idx1 = (idx1+1) % len(cir_arr.cir)
    return True

  # the method take another circular array and returns a linear array containing the common elements between the two circular arrays.
  def intersection(self, c2):
    # To Do
    idx1 = self.start
    count = 0
    temp = [0]*self.size
    for i in range(0,self.size):
      idx2 = c2.start
      for j in range(0,c2.size):
        if self.cir[idx1] == c2.cir[idx2]:
          count += 1
          temp[i] = self.cir[idx1]
        idx2 = (idx2+1) % len(c2.cir)
      idx1 = (idx1+1) % len(self.cir)
    c = [0]*count
    idx = 0
    for i in temp:
      if i != 0:
        c[idx] = i
        idx += 1
    return c
```
